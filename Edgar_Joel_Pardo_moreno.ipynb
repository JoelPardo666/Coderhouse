## Muerte coronaria
La cardiopatía coronaria (CHD), es el término que describe lo que sucede cuando el suministro de sangre del
corazón se bloquea o se interrumpe debido a la acumulación de sustancias grasas en las arterias coronarias. Se
tiene una muestra retrospectiva de hombres en una región de alto riesgo de enfermedad cardíaca de Western
Cape, Sudáfrica. Hay aproximadamente dos controles por cada caso de CHD. Muchos de los hombres con CHD
positivo han sido sometidos a pruebas de sangre, tratamiento de reducción de presión y otros programas para
reducir su riesgo después de su evento de CHD. Estos datos se toman de una base de datos de mayor tamaño.
